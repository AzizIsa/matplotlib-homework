```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_data= pd.read_csv(city_data_to_load)
ride_data= pd.read_csv(ride_data_to_load)

# Combine the data into a single dataset
new_df = pd.merge(ride_data, city_data, on="city", how="left" )

# Display the data table for preview
new_df.head()
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types
city_types = new_df.groupby(["city","type"])
total_rides = city_types["ride_id"].count()
average_fare = city_types["fare"].mean()
driver_count = city_types["driver_count"].sum()
df= pd.DataFrame({"average fare": average_fare, "total rider":total_rides, "total driver": driver_count})
df=df.reset_index()

urban = df.loc[df["type"] == "Urban",:]
urban_x = urban["average fare"]
urban_y = urban["total rider"]
urban_z = urban["total driver"]

rural = df.loc[df["type"] == "Rural",:]
rural_x = rural["average fare"]
rural_y = rural["total rider"]
rural_z = rural["total driver"]

suburban=df.loc[df["type"] == "Suburban",:]
suburban_x = suburban["average fare"]
suburban_y = suburban["total rider"]
suburban_z = suburban["total driver"]


# Build the scatter plots for each city types

plt.scatter( urban_y, urban_x, urban_z,  marker="o", c='lightcoral', edgecolors="black", alpha=0.75, label="Urban")
plt.scatter(rural_y,rural_x,  rural_z, marker="o", c='r', edgecolors="black",  alpha=0.75, label="25Suburban")
plt.scatter(suburban_y, suburban_x,  suburban_z, marker="o", c='b', edgecolors="black",  alpha=0.75, label="Rural")


# Incorporate the other graph properties

# Create a legend

# Incorporate a text label regarding circle size

# Save Figure

#Buble sizes
# rural_bubble = rural['total driver'] * 10
# suburban_bubble = suburban['total driver'] * 10
# urban_bubble = urban['total driver'] * 10
```
